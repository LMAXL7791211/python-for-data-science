В каждом графике массив x используется для координат по горизонтали.

Дайте название графикам: 'График y1', 'График y2' и т.д.

Для графика в левом верхнем углу установите границы по оси x от -5 до 5.
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y1
```

```python
y2 = 2 * x + 0.5
y2
```

```python
y3 = -3 * x - 1.5
y3
```

```python
y4 = np.sin(x)
y4
```

```python
title_dict = {'fontsize': 10, 'fontweight': 'bold', 'color': '#808080'}
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
xy1, xy2, xy3, xy4 = ax.flatten()

xy1.plot(x, y1)
xy1.set_title('График y1', fontdict=title_dict)
xy1.set_xlim([-5, 5])
xy2.plot(x, y2)
xy2.set_title('График y2', fontdict=title_dict)
xy3.plot(x, y3)
xy3.set_title('График y3', fontdict=title_dict)
xy4.plot(x, y4)
xy4.set_title('График y4', fontdict=title_dict)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
